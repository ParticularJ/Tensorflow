```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#
X = np.c_[.5,1].T
print(X)
y = [.5, 1]
test = np.c_[0, 2].T
regr = linear_model.Ridge(alpha=.1)
```

```python
plt.figure()
```

```python
np.random.seed(0)
```

```python
for _ in range(10):
    this_X = .1 * np.random.normal(size=(2, 1)) + X
    print(this_X)
    regr.fit(this_X, y)
    plt.plot(test, regr.predict(test))
    plt.scatter(this_X, y, s=3)
```

```python
from sklearn import datasets, neighbors, linear_model
```

```python
digits = datasets.load_digits()
digits.data.shape
```

```python
X_digits = digits.data[:-100]
X_digits.shape
```

```python
y_digits = digits.target[:-100]
```

```python
knn = neighbors.KNeighborsClassifier()
```

```python
knn.fit(X_digits, y_digits)
```

```python
a=knn.predict(digits.data[-10:])
```

```python
knn.score(digits.data[-10:], digits.target[-10:])
```

```python
logistic = linear_model.LogisticRegression()
```

```python
logistic.fit(digits.data[:-1000], digits.target[:-1000])
```

```python
logistic.score(digits.data[-10:], digits.target[-10:])
```
